```python
from tqdm.notebook import tqdm
from ipywidgets import FloatProgress
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
scraper=sntwitter.TwitterSearchScraper("capres 2024")
tweets=[]
n_tweets=500
for i, tweet in tqdm(enumerate(scraper.get_items()), total=n_tweets):
    data=[
        tweet.user.username, 
        tweet.rawContent, 
        tweet.user.location,
    ]
    tweets.append(data)
    if i > n_tweets:
        break
tweet_df=pd.DataFrame(tweets, columns=['@Username','Tweet', 'Lokasi'])
tweet_df
tweet_df.to_csv('Data_Twitter.csv', index=False)
```
