# Task 8 (PageRank)

PageRank adalah algoritma yang dibuat oleh Larry Page. Algoritma ini digunakan oleh mesin pencarian Google yang memberikan bobot numerik untuk setiap elemen dari kumpulan dokumen hyperlink seperti Word Wide Web dengan tujuan untuk mengukur hubungan kepentingan dalam kumpulan dokumen tersebut.

```python
import networkx as nx
import pylab as plt
# membuat graoh berarah
G = nx.DiGraph()

# membuat node 1-9
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8, 9])
# membuat panah graph berarah
G.add_edges_from([(1, 2), (1, 3), (1, 4), (2, 5), (2, 6), (3, 5), (3, 6), (4, 6), (5, 7), (6, 7), (7, 8), (8, 9), (9, 7)])
nx.draw(G, with_labels = True)
```

```python
# melakukan kalkulasi pagerank dengan menetapkan nilai Damping factor = 0.85
d = 0.85
pagerank = {}

# melakukan inisiasi pagerank masing-masing nilai node 1/N
N = len(G.nodes())
for node in G.nodes():
    pagerank[node] = 1/N

# melakukan iterasi graph untuk 100-1 iterasi
for i in range(100):
    # update nilai pagerank untuk setiap node
    for node in G.nodes():
        new_pagerank = (1 - d)/N
        for predecessor in G.predecessors(node):
            num_successors = len(list(G.successors(predecessor))) #menghitung jumlah outbound link dari masing2 node yg mengacu pada node yg akan dihitung
            new_pagerank += d * pagerank[predecessor] / num_successors #mengalikan damping factor dengan total seluruh nilai pagerank dibagi dengan masing2 nilai outbound linknya
        pagerank[node] = new_pagerank

# print pagerank values
print("Pagerank values:")
for node in pagerank:
    print(f"Node {node}: {pagerank[node]}")
```
